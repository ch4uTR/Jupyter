```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../datasets/heart_attack_dataset.csv")
df.head(3)
```

```python
df.info()
```

```python
y = df["target"]
x = df.drop("target", axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 57, train_size = 0.75)
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
scaled_model = DecisionTreeClassifier()
scaled_model.fit(x_train_scaled, y_train)
```

```python
scaled_model.score(x_test_scaled, y_test)
```

```python
dot_data = export_graphviz(model, filled = True, rounded = True)
graph = graphviz.Source(dot_data)
```
