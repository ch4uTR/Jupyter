```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv("../datasets/Student_Grades.csv")
df.head(3)
```

```python
y = df["Scores"]
x = df.drop(columns = ["Grade", "Scores"])
```

```python
df.info()
```

```python
y = y.values.astype("float32")
x = x.values.astype("float32")
```

```python
model = Sequential()

model.add(Dense(16,activation = "relu"))
model.add(Dense(32, activation = "relu"))
model.add(Dense(32, activation = "relu"))
model.add(Dense(1))

model.compile(optimizer = "adam", loss = "mse")
```

```python
model.fit(x,y, epochs = 150, batch_size = 16)
```

```python
model.predict(np.array([[2.5, 0.5, 0, 2, 2.3]]))
```

```python
model.predict(np.array([[5.1, 1.5, 0.5, 4, 4.9]]))
```
