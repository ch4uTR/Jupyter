```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../datasets/Audi_A1_listings.csv")
df.head()
```

```python
df. info()
```

```python
cols = ["index", "href", "PPY", "MileageRank", "PriceRank", "PPYRank", "Score"]
df = df.drop(columns=cols, axis=1)
df.info()
```

```python
df.head(3)
```

```python
df["Engine"] = df["Engine"].str.replace("L", "").astype(float)
df.head(3)
```

```python
cat_cols = ["Type", "Transmission", "Fuel"]
df[cat_cols].nunique()
```

```python
df[cat_cols].head()
```

```python
df = pd.get_dummies(df, columns = cat_cols, drop_first=True, dtype=int) 
df.head(3) 
```

```python
df.tail()
```

```python
df.info()
```

```python
y, x = df[["Price(£)"]], df.drop(columns=["Price(£)"],axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=57, train_size=0.75)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
